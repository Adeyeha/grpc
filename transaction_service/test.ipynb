```python
import grpc
import random
from transaction_proto import transaction_pb2_grpc,transaction_pb2



with grpc.insecure_channel('localhost:50053') as channel:
    response = stub = transaction_pb2_grpc.TransactionControllerStub(channel)
    # print('----- Destroy All ------')
    # stub.DestroyAll(transaction_pb2.Transaction())
    print('----- Create -----')
    response = stub.Create(transaction_pb2.Transaction(amount=round(random.uniform(1, 10000),2), currency='EUR', dc_ind='Debit', description='Transfer to Family', account='1000000001'))
    print(response, end='')
    print('----- List -----')
    for post in stub.List(transaction_pb2.TransactionListRequest()):
        print(post, end='')
    print('----- Retrieve By Account-----')
    for post in stub.RetrieveByAccount(transaction_pb2.TransactionRetrieveByAccountRequest(account=response.account)):
        print(post, end='')
    print('----- Retrieve -----')
    response = stub.Retrieve(transaction_pb2.TransactionRetrieveRequest(transaction_id=response.transaction_id))
    print(response, end='')
    print('----- Update -----')
    response = stub.Update(transaction_pb2.Transaction(transaction_id=response.transaction_id, amount=round(random.uniform(1, 10000),2), currency='CAD', dc_ind='Debit', description='Transfer to Family', account=response.account))
    print(response, end='')
    # print('----- Delete -----')
    # stub.Destroy(transaction_pb2.Transaction(transaction_id=response.transaction_id))
```
